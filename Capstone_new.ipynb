```python
def f(x):
    if (x > 4) and (x <= 8):
        return 0
    elif (x > 8) and (x <= 12 ):
        return 1
    elif (x > 12) and (x <= 16):
        return 2
    elif (x > 16) and (x <= 20) :
        return 3
    elif (x > 20) and (x <= 24):
        return 4
    elif (x <= 4):
        return 5
```

```python
# ship daypart as numbers , basically making a category
df1['ship_daypart'] = df1['ship_hour'].apply(f)
#order daypart
df1['order_daypart'] = df1['order_hour'].apply(f)
df1.head()
```

```python
import pandas as pd
from scipy import stats

group_0 = final_df[final_df['Late_delivery_risk'] == 0]['Days for shipment (scheduled)']
group_1 = final_df[final_df['Late_delivery_risk'] == 1]['Days for shipment (scheduled)']

# Perform a t-test to compare the means of the two groups
t_stat, p_value = stats.ttest_ind(group_0, group_1, equal_var=False)  # Welch's t-test (does not assume equal variance)

# Output the test statistic and p-value
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05  # Common significance level

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in 'Days for shipment (scheduled)' between the two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in 'Days for shipment (scheduled)' between the two groups.")
```

```python
score_card = pd.DataFrame(columns=['Model Name', 'Train Accuracy Score','Test Accuracy Score','f1-micro',
                                       'f1-macro', 'f1-weighted'])


def update_score_card(model_name,model,X_train,X_test,y_train,y_test):
    from sklearn import metrics
    y_pred_train = model.predict(X_train)
    y_pred = model.predict(X_test)
    
 
    global score_card

    
    score_card=pd.concat([score_card,pd.DataFrame({'Model Name':[model_name],
                                    'Train Accuracy Score': [metrics.accuracy_score(y_train, y_pred_train)],
                                    'Test Accuracy Score': [metrics.accuracy_score(y_test, y_pred)],
                                    'f1-micro':[metrics.f1_score(y_test,y_pred,average='micro')],
                                    'f1-macro':[metrics.f1_score(y_test,y_pred,average='macro')],
                                    'f1-weighted':[metrics.f1_score(y_test,y_pred,average='weighted')]})])
    return score_card
```

```python
abcl = AdaBoostClassifier(n_estimators = 6,random_state = 10)
abcl.fit(xtrain,ytrain)
update_score_card('AdaBoostClassifier',abcl,xtrain,xtest,ytrain,ytest)
```

```python
gbcl = GradientBoostingClassifier(n_estimators = 5,random_state = 10)
gbcl.fit(xtrain,ytrain)
update_score_card('GradientBoostingClassifier',gbcl,xtrain,xtest,ytrain,ytest)
```

```python
dc = DecisionTreeClassifier()
dc.fit(xtrain,ytrain)
update_score_card('DecisionTreeClassifier',dc,xtrain,xtest,ytrain,ytest)
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
update_score_card('LogisticRegression',model,xtrain,xtest,ytrain,ytest)
```
